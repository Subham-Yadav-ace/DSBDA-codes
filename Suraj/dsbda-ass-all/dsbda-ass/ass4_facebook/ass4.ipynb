```python
# --- a. Create Data Subsets ---
# Subset by rows and columns using iloc

subset_row_col = fb.iloc[2:10,1:5]
subset_row_col
```

```python
# Subset by specific column names
subset_columns = fb[['Type', 'like', 'comment']]
subset_columns
```

```python
# Subset based on a condition (filtering)
# Creating 'Z' as seen in your reference: likes > 100 and share > 40
Z = fb[(fb['like'] > 100) & (fb['share'] > 40)]
Z
```

```python
# --- b. Merge Data ---
# Merging the two dataframes vertically using concat (replaces deprecated .append)
merged = pd.concat([fb, newfb])
merged
```

```python
# --- c. Sort Data ---
# Sorting the subset Z by the 'comment' column in descending order
sorted_data = Z.sort_values(by='comment', ascending=False)
sorted_data
```

```python
# --- d. Transposing Data ---
# Flipping rows and columns
transposed_data = Z.T
transposed_data
```

```python
# --- e. Shape and Reshape Data ---
# Check dimensions
print(f"Original Data Shape: {fb.shape}")
print(f"Merged Data Shape: {newfb.shape}")

# Reshaping data using melt (Turning wide format to long format)
reshaped_data = Z.melt(id_vars=['Type'], value_vars=['like', 'comment'])
reshaped_data
```

```python
# --- Extra: Exporting Data ---
# Saving the processed subset to files as per your notebook
Z.to_csv('output.csv', index=False)
Z.to_html('output.html', index=False)
```
