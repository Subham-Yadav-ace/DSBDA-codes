```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("heart.csv")
df
```

```python
print(df.head())
print(df.shape)
print(df.columns)
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
df
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df.corr(),annot=True,cmap="coolwarm")
plt.title("Correlation Between Features")
plt.show()
```

```python
sns.lineplot(data=df,x='age',y='chol',hue='target')
plt.title("Age vs Cholestrol")
plt.show()

sns.lineplot(data=df, x='age',y='trestbps',hue='target')
plt.title("Age vs Blood Pressure")
plt.show()
```

```python
# sns.lineplot(data=df, x='age',y='trestbps',hue='target')
sns.histplot(data=df, x='age',hue='target',kde=True)
plt.title("Age vs Blood Pressure")
plt.show()

sns.histplot(data=df, x='cp',hue='target',kde=True)
plt.title("Age vs Blood Pressure")
plt.show()

sns.histplot(data=df, x='target',hue='sex')
plt.title("Age vs Blood Pressure")
plt.show()
```

```python
subset = df[['age','cp','trestbps','chol','oldpeak','target']]
sns.pairplot(subset,hue='target')
plt.show()
```

```python
plt.figure(figsize=(15,10))

i = 1
for col in subset.columns:
    plt.subplot(3, 3, i)
    sns.histplot(subset[col], kde=True)
    plt.title(f"Distribution of {col}")
    i += 1

plt.tight_layout()
plt.show()
```
