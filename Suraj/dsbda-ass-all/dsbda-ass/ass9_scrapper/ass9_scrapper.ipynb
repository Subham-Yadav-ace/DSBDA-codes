```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Change the URL to scrape a different website

URL = "https://books.toscrape.com"

# HEADERS = {
#     "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) Chrome/124.0.0.0 Safari/537.36"
# }

# HEADERS = {
#     "User-Agent": "Mozilla/5.0"
# }
```

```python
#Sending request to website
response = requests.get(URL)

print(f"Status Code: {response.status_code}")

if(response.status_code == 200):
    print("Page Fetched Successfully...")
else:
    print("Failes to fetch page. Check the URL...")
```

```python
#Parse HTML
soup = BeautifulSoup(response.text,"lxml")

#Checking the Parsing is Successful
print("Page Title: ",soup.title.text)
print("HTML Parsed Successfully...")
```

```python
# Using the BeautifulSoup library in Python, you use soup.select() to target these items. 
# The syntax is tag.classname

# Target Tag: article
# Class Name: product_pod
# CSS Selector: article.product_pod

products = soup.select("article.product_pod")
print(f"Found {len(products)} products on this page...")
```

```python
data = []
for product in products:

    # Product Name : name is inside <h3><a title="Book Name">...
    name_tag = product.select_one("h3 > a")
    name = name_tag["title"] if name_tag else "N/A"

    #Price: price is inside <p class="price_color">£12.99
    price_tag = product.select_one("p.price_color")
    price = price_tag.text.strip() if price_tag else "N/A"

    #Availability: availability is inside <p class="instock availability">
    avail_tag = product.select_one("p.availability")
    availability = avail_tag.text.strip() if avail_tag else "N/A"

    #Add this product's data to out list
    data.append({
        "Product Name": name,
        "Price": price,
        "Availability": availability
    })

print(f"Extracted data for {len(data)} products...")
    
for item in data:
    print(item['Product Name'], "|", item['Price'], "|", item['Availability'])
    print()
    
df = pd.DataFrame(data)
df.to_csv("products.csv", index=False)
print("Saved to products.csv")
df
    
```
