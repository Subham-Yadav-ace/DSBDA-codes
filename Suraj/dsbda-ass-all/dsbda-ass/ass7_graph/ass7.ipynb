```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
df

# 1 → person has heart disease
# 0 → person does NOT have heart disease
```

```python
print(df.head())
print(df.shape)
print(df.columns)
```

```python
# 4. Remove duplicates
df = df.drop_duplicates()
df
```

```python
# 5. Check missing values
# print(df.isna().sum())
print(df.isnull().sum())
```

```python
# Remove extra spaces in column names (important fix)
df.columns = df.columns.str.strip()
```

```python
# 7. Correlation Heatmap
#   +1 → strong positive relation (both increase together)
#   -1 → strong negative relation (one increases, other decreases)
#    0 → no relation
# ==============================
plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Features")
plt.show()
```

```python
# 8. Line Plots
# ==============================
sns.lineplot(data=df, x='age', y='chol', hue='target')
plt.title("Age vs Cholesterol")
plt.show()

sns.lineplot(data=df, x='age', y='trestbps', hue='target')
plt.title("Age vs Blood Pressure")
plt.show()
```

```python
# 9. Histograms
# ==============================

# Age distribution
sns.histplot(data=df, x='age', hue='target', kde=True)
plt.title("Age Distribution by Heart Disease")
plt.show()

# Chest pain type
sns.histplot(data=df, x='cp', hue='target')
plt.title("Chest Pain vs Heart Disease")
plt.show()

# Gender vs disease
sns.histplot(data=df, x='target', hue='sex')
plt.title("Heart Disease by Gender")
plt.show()
```

```python
# 10. Pairplot
# ==============================
subset = df[['age','cp','trestbps','chol','oldpeak','target']]
sns.pairplot(subset, hue='target')
plt.show()
```

```python
# 11. Distribution Plots
# ==============================
plt.figure(figsize=(15,10))

for i, col in enumerate(subset.columns, 1):
    plt.subplot(3, 3, i)
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")

plt.tight_layout()
plt.show()
```

```python
group = df.groupby('sex')['target'].sum()

labels = ['Female Disease', 'Male Disease']

plt.pie(group, labels=labels, autopct='%1.1f%%')
plt.title("Heart Disease by Gender")
plt.show()
```
